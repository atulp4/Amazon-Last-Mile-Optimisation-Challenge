# Actual Sequence data of each route and each stop on the route

# Package Data

```python
# Path to the JSON file
package_data_path = '/content/drive/MyDrive/Atul - Optimization and Routing/data/model_build_inputs/package_data.json'

# Load JSON data
with open(package_data_path, 'r') as file:
    package_data = json.load(file)

# Convert to DataFrame
package_data_list = []
for route_id, stops in package_data.items():
    for stop_id, packages in stops.items():
        for package_id, details in packages.items():
            package_data_list.append({
                'RouteID': route_id,
                'StopID': stop_id,
                'PackageID': package_id,
                'ScanStatus': details['scan_status'],
                'StartTime': details['time_window'].get('start_time_utc'),
                'EndTime': details['time_window'].get('end_time_utc'),
                'ServiceTime': details['planned_service_time_seconds'],
                'Depth': details['dimensions']['depth_cm'],
                'Height': details['dimensions']['height_cm'],
                'Width': details['dimensions']['width_cm']
            })

package_data_df = pd.DataFrame(package_data_list)
print(package_data_df.head())

# Save DataFrame to CSV
package_data_df.to_csv('/content/drive/MyDrive/Atul - Optimization and Routing/data/package_data.csv', index=False)
```

# Route & Stop Data

```python
# Path to the JSON file
route_data_path = '/content/drive/MyDrive/Atul - Optimization and Routing/data/model_build_inputs/route_data.json'

# Load JSON data
with open(route_data_path, 'r') as file:
    route_data = json.load(file)

# Convert to DataFrame
route_data_list = []
stop_data_list = []
for route_id, details in route_data.items():
    route_data_list.append({
        'RouteID': route_id,
        'StationCode': details['station_code'],
        'Date': details['date_YYYY_MM_DD'],
        'DepartureTime': details['departure_time_utc'],
        'Capacity': details['executor_capacity_cm3'],
        'RouteScore': details['route_score']
    })
    for stop_id, stop_details in details['stops'].items():
        stop_data_list.append({
            'RouteID': route_id,
            'StopID': stop_id,
            'Latitude': stop_details['lat'],
            'Longitude': stop_details['lng'],
            'Type': stop_details['type'],
            'ZoneID': stop_details['zone_id']
        })

route_data_df = pd.DataFrame(route_data_list)
stop_data_df = pd.DataFrame(stop_data_list)
# print(route_data_df.head())
# print(stop_data_df.head())

# Save DataFrames to CSV
route_data_df.to_csv('/content/drive/MyDrive/Atul - Optimization and Routing/data/route_data.csv', index=False)
stop_data_df.to_csv('/content/drive/MyDrive/Atul - Optimization and Routing/data/stop_data.csv', index=False)
```

# Travel Times

```python
import json
import pandas as pd
import csv

# Path to the JSON file
travel_times_path = '/content/drive/MyDrive/Atul - Optimization and Routing/data/model_build_inputs/travel_times.json'

# Define a function to process the JSON data in chunks
def process_travel_times_chunk(chunk, writer):
    travel_times_list = []
    for route_id, stop_times in chunk.items():
        for stop_id_start, times in stop_times.items():
            for stop_id_end, time in times.items():
                travel_times_list.append({
                    'RouteID': route_id,
                    'StopID_Start': stop_id_start,
                    'StopID_End': stop_id_end,
                    'TravelTime': time
                })

    # Write the chunk to CSV
    df_chunk = pd.DataFrame(travel_times_list)
    df_chunk.to_csv(writer, header=False, index=False)

# Read and process the JSON file in chunks
chunk_size = 1000  # Adjust this size based on your memory constraints
with open(travel_times_path, 'r') as file:
    travel_times_data = json.load(file)

# Open the CSV file for writing
csv_file_path = '/content/drive/MyDrive/Atul - Optimization and Routing/data/travel_times.csv'
with open(csv_file_path, 'w', newline='') as csvfile:
    fieldnames = ['RouteID', 'StopID_Start', 'StopID_End', 'TravelTime']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    chunk = {}
    count = 0
    for route_id, stop_times in travel_times_data.items():
        chunk[route_id] = stop_times
        count += 1
        if count >= chunk_size:
            process_travel_times_chunk(chunk, csvfile)
            chunk = {}
            count = 0

    # Process any remaining data
    if chunk:
        process_travel_times_chunk(chunk, csvfile)

print("Processing complete. CSV file saved.")


```
